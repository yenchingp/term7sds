```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
# Set headers to make request look more like a legitimate user's request
# without this, the server will reject the request
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36 Edge/16.16299',
    'Accept-Language': 'en-US,en;q=0.9'
}
```

```python
base_url = "https://clickproperty.sg/condo-directory/page-"
num_pages = 155 # Change this to the number of pages you want to scrape
```

```python
csv_filename = "condo_data.csv"
csv_file = open(csv_filename, "w", newline="")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Condo Name", "Address", "Postal Code"])
```

```python
#first page


url = f"https://clickproperty.sg/condo-directory"
response = requests.get(url)

if response.no_result_code == 200:
    soup = BeautifulSoup(response.text, "html.parser")
    condos = soup.find_all("div", itemprop="address")

for condo in condos:
    condo_name_tag = condo.find_previous("a", title=True)
    if condo_name_tag:
        condo_name = condo_name_tag["title"]

        address = condo.find("span", itemprop="streetAddress").text.strip()
        postal_code = condo.find("meta", itemprop="postalCode")["content"]

        csv_writer.writerow([condo_name, address, postal_code])
        print(f"Scraped: {condo_name}, {address}, {postal_code}")

    else:
        print(f"Failed to retrieve data from page")
```

```python
#subsequent pages

for page_number in range(2, num_pages + 1):
    url = f"{base_url}{page_number}"
    response = requests.get(url)

    if response.no_result_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        condos = soup.find_all("div", itemprop="address")

    for condo in condos:
        condo_name_tag = condo.find_previous("a", title=True)
        if condo_name_tag:
            condo_name = condo_name_tag["title"]

            address = condo.find("span", itemprop="streetAddress").text.strip()
            postal_code = condo.find("meta", itemprop="postalCode")["content"]

            csv_writer.writerow([condo_name, address, postal_code])
            print(f"Scraped: {condo_name}, {address}, {postal_code}")

        else:
            print(f"Failed to retrieve data from page {page_number}")
```

```python
csv_file.close()

print("Scraping complete. Data saved to condo_data.csv")
```
