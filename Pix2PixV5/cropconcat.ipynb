```python
# Process images based on sorted filenames and split into train and test
for idx in indices:

    area_files = sorted(area_files)
    site_plan_files = sorted(site_plan_files)
    
    area_file = area_files[idx]
    site_plan_file = site_plan_files[idx]

    # Extract image numbers from filenames
    image_number = area_file.split('_')[0]

    # Construct file paths
    area_path = os.path.join(area_folder, area_file)
    site_plan_path = os.path.join(site_plan_folder, site_plan_file)

    # Open and crop the middle 883x883px of the area image
    area_image = Image.open(area_path)
    center_x = area_image.width // 2
    center_y = area_image.height // 2
    half_width = 884 // 2
    half_height = 884 // 2
    area_image = area_image.crop((center_x - half_width, center_y - half_height, center_x + half_width, center_y + half_height))

    # Open and crop the middle 883x883px of the site plan image
    site_plan_image = Image.open(site_plan_path)
    site_plan_image = site_plan_image.crop((center_x - half_width, center_y - half_height, center_x + half_width, center_y + half_height))

    # Resize the images to 512x512
    area_image = area_image.resize((512, 512))
    site_plan_image = site_plan_image.resize((512, 512))

    # Create a new image by concatenating the two resized images into 512x1024
    new_image = Image.new("RGB", (1024, 512))
    new_image.paste(site_plan_image, (0, 0))
    new_image.paste(area_image, (512, 0))

    if image_number not in test_indexes:
        output_path = os.path.join(train_folder, f"{image_number}.png")
    else:
        output_path = os.path.join(test_folder, f"{image_number}.png")

    new_image.save(output_path)
```

```python
# concatenate all images in area_folder and site_plan_folder to 1024x512
for idx in range(len(area_files)):

    area_files = sorted(area_files)
    site_plan_files = sorted(site_plan_files)
    
    area_file = area_files[idx]
    site_plan_file = site_plan_files[idx]

    image_name = str(area_file)

    # Construct file paths
    area_path = os.path.join(area_folder, area_file)
    site_plan_path = os.path.join(site_plan_folder, site_plan_file)

    # Open and crop the middle 883x883px of the area image
    area_image = Image.open(area_path)

    # Open and crop the middle 883x883px of the site plan image
    site_plan_image = Image.open(site_plan_path)

    # Resize the images to 512x512
    area_image = area_image.resize((512, 512))
    site_plan_image = site_plan_image.resize((512, 512))

    # Create a new image by concatenating the two resized images into 512x1024
    new_image = Image.new("RGB", (1024, 512))
    new_image.paste(site_plan_image, (0, 0))
    new_image.paste(area_image, (512, 0))

    output_path = os.path.join(output_folder, f"{image_name}.png")
    new_image.save(output_path)
```
