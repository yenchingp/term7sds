```python
import os

# Paths to the directories containing the datasets
dataset_paths = ['/Users/raymondharrison/Desktop/final_rd/input',
                  '/Users/raymondharrison/Desktop/final_clear/input',
                    '/Users/raymondharrison/Desktop/final_buildings/input',
                    '/Users/raymondharrison/Desktop/final_lines/input',
                    '/Users/raymondharrison/Desktop/final_sat/input',
                    ]

# Function to extract indexes from a given dataset
def get_indexes(dataset_path):
    indexes = set()
    for filename in os.listdir(dataset_path):
        if filename.endswith('.png'):
            index = filename.split('_')[0]
            indexes.add(index)
    return indexes
```

```python

# Extract indexes from each dataset
dataset_indexes = [get_indexes(path) for path in dataset_paths]

# Find indexes present in all datasets
common_indexes = set.intersection(*dataset_indexes)

# Find indexes unique to each dataset
unique_indexes = [indexes - common_indexes for indexes in dataset_indexes]

# Print results
print("Indexes common to all datasets:", common_indexes)
for i, unique in enumerate(unique_indexes, start=1):
    print(f"Indexes unique to dataset {i}:", unique)
```
