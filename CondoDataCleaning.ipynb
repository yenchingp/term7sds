```python
import pandas as pd
```

```python
# read condo_data.csv
pd = pd.read_csv('condo_data.csv')
pd
```

```python
pd = pd[pd['Postal Code'].notna()]
# convert the values in column 3 to str
pd['Postal Code'] = pd['Postal Code'].astype(str)
# remove .0 from the end of teh values in 'Postal Code' column
pd['Postal Code'] = pd['Postal Code'].map(lambda x: x.rstrip('.0'))
# create new column called 'Condo Name No Num' with the values in 'Condo Name' but without any integers
pd['Condo Name No Num'] = pd['Condo Name'].str.replace('\d+', '')
# if there are duplicates in column 'Condo Name No Num', delete that respective row
pd.drop_duplicates(subset='Condo Name No Num', keep='first', inplace=True)
# concat the values in column 3with the values in column 2 with a comma and space in between
pd['Address'] = pd['Address'] + ', ' + pd['Postal Code']

# if the first element in 'Address' is not an integer delete that row
pd = pd[pd['Address'].str.split().str[0].str.isdigit()]

del pd['Postal Code']
del pd['Condo Name No Num']

pd
```

```python
# Save pd as a new csv file
pd.to_csv('condo_data_new.csv', index=False)
```
